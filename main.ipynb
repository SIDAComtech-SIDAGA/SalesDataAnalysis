```python
# EXPLORATION DATA ANALYSIS
# 1.Bar plot to show branch has the highest sales record
# 2.Pie chart to visualize the gender distribution in the data 
# 3.Line chart to analyze sales trend over the TimeoutError
# 4.Histogram to understand which city has the highest sales, differentiated by gender
# 5.Bar chart to determine product line has the most set_matplotlib_close
# 6.Sublots to analyze sales by product line and city, differentiate by gender
# 7.Calculate the avearage rating for each branch and product line
```

```python
# How many record in each brach wise 
dataBranchWise = data['Branch'].value_counts().sort_values(ascending=False)
dataBranchWise.plot(kind='bar', figsize=(10,5),color=['yellow','red','aqua'])
plt.title('Number of records per each branch')
plt.xlabel('Branch')
plt.ylabel('Count of values')
plt.show()
```

```python
#Calculate a pie chart for gender percentages


dataGenderPercentages = data['Gender'].value_counts().sort_values(ascending=False)
dataGenderPercentages.plot(kind='pie',figsize=(10,5),
                           explode=[0,0],
                           labels=['Females','Male'],
                           colors=['red','green'],
                           shadow=True)
plt.title('Gender distribution')
plt.legend()
plt.show()
```

```python
#Create a line chart to understand the sales trend
plt.figure(figsize=(10,5))
sns.lineplot(data, x = 'Date', y = 'Total', palette = 'coolwarm', color = 'red')
plt.title("Every day Total Sales")
plt.xlabel("Date")
plt.ylabel("Total sales amount")
plt.xticks(rotation=90)
plt.show()
```

```python
'''Create hostogram to understand the total sales
done in each state with gender wise using ploty'''

fig = px.histogram(data, x = 'City', y = 'Total', color = 'Gender', title = 'Total sales done')
fig.update_layout(plot_bgcolor = '#90e0ef')
fig.show()
```

```python
#Create a barchart to understand the total sales done in each product wise
colors = ['red','yellow','black','blue','aqua','green','white']
dataProductWise = data.groupby('Product line')['Total'].sum().sort_values(ascending=False)
dataProductWise.plot(kind='bar',title='Total sales done in each product', color=colors,figsize=(10,5))
plt.xlabel('Product')
plt.ylabel('Sum of the sales')
plt.show()
```

```python
# GENDER BEHAVIOUR 
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
sns.barplot(ax=ax[0],data=data,x='City', y='Total', hue='Gender', palette='Oranges')
ax[0].set_title('Total Sales done each city')
plt.xlabel('City')
# plt.legend()
plt.ylabel('Sum of sales')

# Second plt 
sns.barplot(ax=ax[1],data=data, x='Product line', y='Total',hue='Gender', palette='Oranges')
ax[1].set_title('Total Sales done each city')
plt.xlabel('Product line')
plt.ylabel('Total sales')
plt.xticks(rotation=60)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# FIND THE RATINGS 
'''Each branch average rating we use groubby functions and the rating convert into four out of 5 in same we alsoe do some process for '''
branch_rating = pd.DataFrame(data.groupby('Branch')['Rating'].mean().sort_values())
branch_rating
```

```python
# Relationship between unit price and the quantity using scatter plot 
plt.figure(figsize=(10,5))
sns.scatterplot(data=data, x = 'Unit price', y = 'Quantity', color='black')
plt.title('Relationship Between Unit and Quantity Sold')
plt.xlabel('Unit Price')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
#Scatter a scatter plot for total sales vs gross income

plt.figure(figsize=(10,5))
sns.scatterplot(data=data, x='Total', y = 'gross income', color='black')
plt.title('Relationship Between Total sales and gross income')
plt.xlabel('Total Sales')
plt.ylabel('Gross income')
```
